```python
# Ridge Regresyon overfitting durumlarında kullanılır
# Ridge Regresyon sayesinde bias ve varyans arasındaki dengeyi sağlayabiliriz
# Ridge Regresyon'da katsayılar üzerinde regülasyonlar yapılır
# Rigde Regresyon'da katsayılar küçülür ama 0 olmaz
# Ridge Regresyon cezaların karesi ile orantılıdır
# Diğer bir adı L2 Regresyondur
# L1 Regresyon ise katsayıları gerektiğinde 0 yapabilen algoritmadır.
```

```python
# y = a1 * x1 + a2 * x2 + .... + b + alfa * (katsayılar toplamı) **2

# 50 = 40 + 10 + 0
# 50 = 30 + 10 + 10
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge
```

```python
df = pd.read_csv('datasets/Student_Marks.csv')
```

```python
df.head(3)
```

```python
y = df['Marks']
x = df[['time_study']]
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(8,8
))
plt.scatter(x,y)
plt.show() 
#görselleştirme sağladık
```

```python
lr = LinearRegression()
model = lr.fit(x,y)
model.score(x,y)
```

```python
alfalar = [1,10,20,100,200]
for a in alfalar:
  r = Ridge(alpha=a)
  modelr = r.fit(x,y)
  skor = modelr.score(x,y)
  print("Skor",skor)
  print("Katsayı",modelr.coef_)

# alfa değerleri arttıkça diğer ağırlıkların nasıl azaldığının göstergesi
```
